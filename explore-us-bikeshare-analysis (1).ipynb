```python
def get_filters(city,day,month):
    # read our data 
    df=pd.read_csv(city_data[city])
    # transform time or date cols to datetime
    df['End Time']=pd.to_datetime(df['End Time'])
    df['Start Time']=pd.to_datetime(df['Start Time'])
    # extract mask
    mask=(df['End Time'].dt.month==month) & (df['End Time'].dt.day==day)
    df=df[mask]
    return df 
```

```python
def get_filters2():
    print("hello lets explore some US bikeshare data")
    city=input("choose cityname (chicago,newyork,washington): ")
    months=['January','February', 'March','April', 'May','June', 'July','All']
    while True:
        month=input(f"choose from the list which month do you want {months}: ")
        if month.title() in months:
            break
        else:
            print("invalid input strict to the list")
    days=['Sunday', 'Thursday', 'Wednesday', 'Monday', 'Tuesday', 'Friday','Saturday','All']
    while True:
        day=input(f"choose from the list which day do you want {days}: ")
        if day.title() in days:
            break
        else:
            print("invalid input strict to the list")
    return city,month,day
    
```

```python
import matplotlib.pyplot as plt
import time
def time_stats():
    df=load_data()
    start_time=time.time()
    df['Start Time']=pd.to_datetime(df['Start Time'])
    month=df['Start Time'].dt.month_name().mode()
    month=month.to_list()[0]
    day=df['Start Time'].dt.day_name().mode()
    day=day.to_list()[0]
    hour=df['Start Time'].dt.hour.mode()
    hour=hour.to_list()[0]
    print(f"the most frquant month people use bikeshare is {month}")
    print(f"the most frquant day people use bikeshare is {day}")
    print(f"the most frquant hour people use bikeshare is {hour}")
    print("this analysis took %s seconds.",(time.time()-start_time))
    print("Raw data is avaliable would you like to show it..")
    choice=input("enter yes or no : ").lower()
    if choice=='yes':
        row=int(input("enter number of rows: "))
        df.head(row)
    
   
```
